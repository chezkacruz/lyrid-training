<h4>Content Added Over the Years</h4>

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
color = ["#b20710", "#221f1f"]

for i, tempType in enumerate(data['type'].value_counts().index):
    #print(tempType)
    ctr = data[data['type'] == tempType]['year_added'].value_counts().sort_index()
    ax.plot(ctr.index, ctr, color=color[i], label=tempType)
    print(tempType + '\n'+ str(ctr) +'\n')

    
plt.xticks(np.arange(2008, 2021, 1))
ax.set_xlim(2008, 2020)

ax.title.set_text('Content Added Over the Years')
for s in ['top', 'right','bottom','left']:
    ax.spines[s].set_visible(False)

fig.text(0.15,0.8,"Movie", fontweight="bold", fontsize=15, color='#b20710')
fig.text(0.15,0.75,"TV Show", fontweight="bold", fontsize=15, color='#221f1f')

plt.show()
```

From this timeline, we can see that the number of TV shows and movies in Netflix started to grow in 2015. There has been a rapid increase in the number of movies added in Netflix in 2016 and both types of content continuosly grew until 2019. The number of new movies added in 2019 dropped while not much new TV shows has been added.

<h4>Top Categories</h4>

Let's take a look at the Categories in Netflix

```python
from collections import Counter
col = "listed_in"
categories = ", ".join(movies['listed_in']).split(", ")
len(Counter(categories))
```

```python
top20 = Counter(categories).most_common(20)
top20.sort(key=lambda x: x[1], reverse=True)
top20
```

```python
labels = [_[0] for _ in top20][::1]
values = [_[1] for _ in top20][::1]
```

```python
bar = sns.barplot(x=values, y=labels)
bar.set_title('All Time Top 20 Categories')
sns.set(rc={'figure.figsize':(10,11)})
```

Here are the categories in Netflix in descending order. As we can see, International Movies and Dramas dominate the categories.

<h4>Top Countries in Netflix</h4>

To get the countries with most content, let us disregard shows without a country listed. Since some shows have multiple countries listed, we will only be considering the first one in the list.


```python
pd.options.mode.chained_assignment = None

data['count'] = 1
df = data.dropna(subset=['country'])
df['count'] = 1
df['first_country'] = df['country'].apply(lambda x: x.split(",")[0])
df.tail()
```

```python
df['count'] = 1
data['count'] = 1

df = df.groupby('first_country')['count'].sum().sort_values(ascending=False)[:10]

fig, ax = plt.subplots(1,1, figsize=(12, 6))
ax.bar(df.index, df, width=0.5, linewidth=0.6)
ax.set_title('Top 10 Content-creating Countries')
ax.set_ylabel('Number of Content In Netflix')
ax.set_xlabel('Country')

plt.show()
```

```python
df
```

As we can see, contents from USA dominate the shows in Netflix. There is a huge gap between USA and India, the country with the second most number of content in the streaming platform.

<h4>Content Added Per Month</h4>

If you want to add a content in Netflix, which month is best to add it?

```python
col = 'month_added'
temp = data[col].value_counts().reset_index()
temp = temp.rename(columns = {col : "count", "index" : col})
temp['percent'] = temp['count'].apply(lambda x : 100*x/sum(temp['count']))
temp["month_added"]

bar = sns.barplot(x=temp["month_added"], y=temp['count'])
bar.set_title('Content Added Per Month')
bar.set_xlabel('Month Added (MM)')
sns.set(rc={'figure.figsize':(8,5)})
```

```python
temp
```

From above, we can see that a lot of new content are added during the months of October and December while not as much are being added on February. Adding your content in February may be beneficial for you since not much contents are being added in this month, so more chances for people looking for new shows to watch may view your show. However, there is no guarantee that your content will gain popularity as other factors may affect how viewers choose to watch your show.

<h4>Netflix titles</h4>

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import random
from PIL import Image
import matplotlib

# Custom colour map based on Netflix palette
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ['#221f1f', '#b20710'])

text = str(list(data['title'])).replace(',', '').replace('[', '').replace("'", '').replace(']', '').replace('.', '')

wordcloud = WordCloud(background_color = 'white', width = 500,  height = 200,colormap=cmap, max_words = 150).generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

From this wordcloud, we can see that titles of movies and tv shows in Netflix share a keyword. We can see that there are a lot of content with the words Love and Christmas. Surprisingly, despite 'Love' being a keyword in the title of a lot of Netflix shows, not much content are being added in the month of February from our analysis a while ago. On the other hand, 'Christmas' is a keyword for a lot of Netflix shows, which is reasonable since a lot of content are being added in the month of December. The keywords 'Time', 'Life', 'World', 'Man' are also common among Netlfix titles. 

Since these words are common among titles, this may imply that a lot of people are interested in watching movies whose plot revolves around these keywords.
